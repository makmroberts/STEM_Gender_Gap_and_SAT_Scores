```python
# merge all dataframes

SAT_data = pd.concat([df1, df2, df3, df4, df5, df6, df7])

SAT_data = SAT_data.reset_index()
SAT_data = SAT_data.drop(columns = ['index'])

SAT_data.head(10)
```

```python
student_numbers = SAT_data[SAT_data['Score Range'] == 'Test Takers'].reset_index().drop(columns = ['Score Range', 'Section','index'])

# Fill NaN values with 0 and convert to integers
columns_to_convert = ['Total', 'Female', 'Male', 'American Indian', 'Asian', 'African American', 'Hispanic', 'Native Hawaiian', 'White', 'More Races']
student_numbers.loc[:,columns_to_convert] = student_numbers.loc[:,columns_to_convert].apply(lambda x: x.str.replace(',', '').fillna(0).astype(int))

student_numbers
```

```python
sat_data_clean = SAT_data[SAT_data['Score Range'] != 'Test Takers'].reset_index().drop(columns = ['index'])

# Remove '%' and convert to int for the specified columns
columns_to_clean = ['Total', 'Female', 'Male', 'American Indian', 'Asian', 'African American', 'Hispanic', 'Native Hawaiian', 'White', 'More Races']
sat_data_clean[columns_to_clean] = sat_data_clean[columns_to_clean].apply(lambda x: x.str.replace('%', '').astype(int)/100)

# convert scores to total student numbers instead of percentages where the year matches
for year in student_numbers['Year']:
    mask = sat_data_clean['Year'] == year
    sat_data_clean.loc[mask, columns_to_clean] = sat_data_clean.loc[mask, columns_to_clean].multiply(student_numbers[student_numbers['Year'] == year][columns_to_clean].values[0], axis=1)

warnings.filterwarnings('ignore')

sat_data_clean
```

```python
sat_tidy = sat_data_clean.melt(id_vars=['Year', 'Section', 'Score Range'], var_name='Demographic', value_name='Students')

sat_tidy
```

```python
# plot average difference highest percentiles in erw and math scores by gender for 2023

top_scores = sat_tidy['Score Range'].unique()[[6,7]]
differences_list = []

for year in years:
    top_subset = sat_tidy[(sat_tidy['Year'] == year) & (sat_tidy['Section'] != 'Total') & (sat_tidy['Score Range'].isin(top_scores)) & (sat_tidy['Demographic'].isin(['Male', 'Female']))]
    
    top_differences = top_subset.groupby(['Section', 'Demographic'])['Students'].sum().unstack('Section')
    top_differences['Difference'] = top_differences['ERW'] - top_differences['Math']
    top_differences = top_differences.reset_index()
    top_differences.drop(columns=['ERW', 'Math'], inplace=True)
    top_differences['Year'] = year
    
    differences_list.append(top_differences)

compiled_differences = pd.concat(differences_list).reset_index(drop=True)
compiled_differences
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

# Increase font sizes
plt.rcParams.update({'font.size': 17})

# Calculate the proportion of the difference relative to the total number of students
compiled_differences = compiled_differences.merge(student_numbers[['Year', 'Total']], on='Year', suffixes=('', '_y'))
compiled_differences['Proportion Difference'] = compiled_differences['Difference'] / compiled_differences['Total'].astype(float)

sns.lineplot(data=compiled_differences, x='Year', y='Proportion Difference', hue='Demographic', marker='o', palette={'Male': 'blue', 'Female': 'red'}, ax=ax)
ax.set_title('Difference in Highest ERW and Math Scores by Gender')
ax.set_ylabel('Difference (ERW - Math)')
ax.set_xlabel('')
ax.axhline(0, color='gray', linestyle='--')
ax.xaxis.label.set_size(17)
ax.yaxis.label.set_size(17)
ax.legend(title='Demographic')
plt.show()
```

```python
# Calculate the average scores by gender and section over the years
average_scores = sat_tidy.groupby(['Year', 'Section', 'Demographic', 'Score Range'])['Students'].mean().reset_index()

# Convert 'Year' to integer
sat_tidy['Year'] = sat_tidy['Year'].astype(int)
average_scores['Year'] = average_scores['Year'].astype(int)

# Filter to include only 'Male' and 'Female' in the 'Demographic' column
average_scores = average_scores[average_scores['Demographic'].isin(['Male', 'Female'])]

# Convert 'Score Range' to numeric by taking the average of the range
average_scores['Score Range'] = average_scores['Score Range'].apply(lambda x: (int(x.split('–')[0]) + int(x.split('–')[1])) / 2)# if '–' in x else int(x.split('–')[0]))

# Calculate the weighted average score
average_scores['Weighted Score'] = average_scores['Score Range'] * average_scores['Students']
average_scores = average_scores.groupby(['Year', 'Section', 'Demographic']).apply(lambda x: x['Weighted Score'].sum() / x['Students'].sum()).reset_index(name='Average Score')

# Separate the data for Math and ERW sections
average_scores_math = average_scores[average_scores['Section'] == 'Math']
average_scores_erw = average_scores[average_scores['Section'] == 'ERW']

# Plot the bar charts
fig, axes = plt.subplots(1, 2, figsize=(18, 8))

# Increase font sizes
plt.rcParams.update({'font.size': 17})

sns.barplot(ax=axes[0], data=average_scores_math, x='Year', y='Average Score', hue='Demographic', ci=None, palette={'Male': 'blue', 'Female': 'red'})
axes[0].set_title('Average Math Scores by Gender')
axes[0].set_ylabel('Average Score')
axes[0].set_xlabel('')
axes[0].set_ylim(400, 600)
axes[0].xaxis.label.set_size(17)
axes[0].yaxis.label.set_size(17)
axes[0].legend(title='Gender')

sns.barplot(ax=axes[1], data=average_scores_erw, x='Year', y='Average Score', hue='Demographic', ci=None, palette={'Male': 'blue', 'Female': 'red'})
axes[1].set_title('Average ERW Scores by Gender')
axes[1].set_ylabel('Average Score')
axes[1].set_xlabel('')
axes[1].set_ylim(400, 600)
axes[1].xaxis.label.set_size(17)
axes[1].yaxis.label.set_size(17)
axes[1].legend(title='Gender')

plt.tight_layout()
plt.show()
```
